```python
from sklearn.preprocessing import normalize as nm
```

```python
import pandas as p

data = p.read_csv("titanic.csv")
data = data.dropna()
```

```python
data_age = data["Age"].values.reshape(-1,1)
data_fare = data["Fare"].values.reshape(-1,1)
```

```python
data_age_norm = nm(data_age)
data_fare_norm = nm(data_fare)
```

```python
data_age_norm
```

```python
data_fare_norm
```

```python
data["Age"] = data_age_norm
data["Fare"] = data_fare_norm
data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
data_age = data["Age"].values.reshape(-1,1)
data_fare = data["Fare"].values.reshape(-1,1)
```

```python
data_age_norm = scaler.fit(data_age)
data_fare_norm = scaler.fit(data_fare)
```

```python
data["Age"] = data_age_norm
data["Fare"] = data_fare_norm
```
