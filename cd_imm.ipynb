```python
data =p.read_excel("Canada.xlsx", sheet_name="Canada by Citizenship", skiprows=range(20), skipfooter=2)
# data =p.read_excel("Canada.xlsx",sheet_name="Canada by Citizenship")
```

```python
# Remove columns -  DEV REG AREA Type Coverage
data.drop(columns=['DEV', 'REG', 'AREA', 'Type', 'Coverage'],inplace=True)
```

```python
data.columns.values
```

```python
# Rename columns - AreaName to Continent, OdName to Country
data.rename(columns={"AreaName" : "Continent", "OdName" : "Country"},inplace=True)
```

```python
# Display record of countries with 'G', P'
data[data['Country'].str.startswith('G')]['Country'].unique()
```

```python
# Display record of countries with 'G', P'
data[data['Country'].str.endswith('G')]['Country'].unique()
```

```python
# Display record of countries with 'G', P'
regex = "[GP]"
data[data['Country'].str.contains(regex)]['Country'].unique()
```

```python
# Find no. of countries which have immigration more than 10000 in year 2009
data[data[2009] > 10000].Country.count()
```

```python
#Find top 5 countries in immigration count for years 2010 to end year
data.sort_values(list(range(2010,2014)),ascending=False).Country.head(5)
```
